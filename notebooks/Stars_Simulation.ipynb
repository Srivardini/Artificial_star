# **Variable Stars**

```python
def create_simulated_event_star_table(n_sources, params=None):
  # Table:
  if params is None:
    param_ranges = {'flux': [1e3, 1e4],
                        'x_mean': [50, 450],
                        'y_mean': [50, 450],
                        'x_stddev': [3.,3.],
                        'y_stddev': [3,3],
                        'theta': [0, np.pi]} 
  else:
    param_ranges = params

  fwhm = param_ranges['x_stddev'][0]*gaussian_sigma_to_fwhm

  df = make_random_gaussians_table(n_sources, param_ranges)

  df.remove_column('amplitude')

  return df,fwhm

def create_simulated_images(shape, df, sky = 50, DC = 3, bias = 10, RN = 3,
                            add_noise = False):

  image = make_gaussian_sources_image(shape, df)
  image += sky

  if not add_noise:
    return image 

  # Source Shot Noise
  image = np.random.poisson(lam = image, size = shape).astype(np.float64)

  # Dark Current
  DC = np.random.poisson(lam = DC, size = shape).astype(np.float64)

  # Bias and Read Noise
  noise = np.random.normal(loc = bias, scale = np.sqrt(RN), size = shape)

  return image + DC + noise

def detect_stars_in_images(image, fwhm):

  mean, median, std = sigma_clipped_stats(image, sigma=3.0)  

  daofind = DAOStarFinder(fwhm=fwhm, threshold=5.*std)  
  sources_stars = daofind(image - median)  

  positions = np.transpose((sources_stars['xcentroid'], 
                            sources_stars['ycentroid']))
  
  apertures = CircularAperture(positions, r=2.0*fwhm)

  photo_table = aperture_photometry(image, apertures)

  return photo_table, apertures
```

```python
df,fwhm = create_simulated_event_star_table(5)

t = np.linspace(0,2*np.pi,5)
fluxes = 4000 + 200*np.sin(t)

shape = (500, 500)
sim_imgs = []
phot_tables = []
apers = []

for i in range(len(fluxes)):

  df[0]['flux'] = fluxes[i]
  img = create_simulated_images(shape, df, add_noise = True)
  phot_table, apertures = detect_event_in_images(img, fwhm)

  sim_imgs.append(img)
  phot_tables.append(phot_table)
  apers.append(apertures)
```

```python
fig, ax = plt.subplots(1,5, figsize = (50,10))
for i in range(5):
  ax[i].imshow(sim_imgs[i], cmap = 'jet', norm = col.LogNorm())
```
